```python
#survival count
sns.countplot(x='Sex',hue='Survived',data=df)
plt.title('Survival Count')
plt.show()
```

```python
#Survival by Gender
sns.countplot(x='Sex',hue='Survived',data=df)
plt.title("Survival by gender")
plt.show()
```

```python
#survival by Pclass
sns.countplot(x='Pclass', hue='Survived',data=df)
plt.title('Survived by passenger class')
plt.show()
```

```python
#Age distribution
plt.figure(figsize=(10,6))
sns.histplot(df['Age'].dropna(),bins=30,kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
#Boxplot Age vs Pclass
sns.boxplot(x='Pclass',y='Age',data=df)
plt.title('Age vs Passenger Class')
plt.show()
```

```python
#Heatmap Correlation Matrix
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(),annot=True,cmap='coolwarm',fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Pairplot (small sample to avoid overload)
sample_df = df[['Survived','Pclass','Age','Fare']].dropna()
sns.pairplot(sample_df,hue='Survived')
plt.title('Pairplot of selected Features',y=1.02)
plt.show()
```
